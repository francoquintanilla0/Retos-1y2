Importamos las librerías necesarias para nuestro dataframe (pandas), gráficas (seaborn) y modelos de aprendizaje (sklearn). 

```python
import pandas as pd 
import seaborn as sns
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score, mean_squared_error, confusion_matrix, accuracy_score, f1_score, mean_absolute_error
```

Cargar el data frame y visualizarlo.

```python
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Documentos IA/Real State/data.csv') 
df.head()
```

Buscamos valores nulos y vemos el tipo de datos en cada columna para saber si es necesario realizar alguna modificación en el dataframe. 

```python
df.info() 
```

Eliminamos los datos nulos para no tener problemas al momento de implementar nuestros modelos de aprendizaje. 

```python
df=df.dropna()
df.info()
```

Realizamos una matriz de correlación para que eliamos las variables con las que utilizaremos para los modelos de aprendizaje dependiendo de su índice de correlación.

```python
corr_df=df.corr(method='pearson')

corr_df.style.background_gradient(cmap='coolwarm')
```

Decidimos definir la variable dependiente a predecir como "TAX" que se refiere a los impuestos de la vivienda.

```python
corr_df['TAX'].sort_values(ascending=True)
```

Eligiremos solo las variables que tengan una correlación moderada que se considera arriba de 0.45 y dividimos los datos en 75% entrenamiento y 25% para pruebas

```python
testp = 0.25
x=df[["CRIM","INDUS","NOX","AGE","RAD","DIS","LSTAT","MEDV"]]
y=df[["TAX"]]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=testp,random_state=1800)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.7)
```

# Bosques aleatorios

```python
#bosque
myForest= ensemble.RandomForestClassifier()
myForest.fit(x_train,y_train)
```

Obtener coeficientes para probar el modelo.

```python
CoD = r2_score(y_test, myForest.predict(x_test))
y_pred = myForest.predict(x_test)
print(
  'mean_squared_error : ', mean_squared_error(y_test, y_pred))
print(
  'mean_absolute_error : ', mean_absolute_error(y_test, y_pred))
print("Score Bosque: ",myForest.score(x_test,y_test))
print("r^2 Bosque: ",CoD)
```

# Regresión Lineal 

Entrenamos el modelo de regresión lineal, para comparar

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

Obtener coeficientes para probar el modelo.

```python
print(
  'mean_squared_error : ', mean_squared_error(y_test, y_pred))
print(
  'mean_absolute_error : ', mean_absolute_error(y_test, y_pred))

print("Score Regresión: ",model.score(x_test,y_test))
CoD = r2_score(y_test,y_pred)
print("r^2 Regresión: ",CoD)
```

# Conclusión


Se obtuvieron mejores resultados con el bosque de decisiones ya que se obtuvo menor error cuadrático y un puntaje y r^2 mayor que en la regresión lineal, esto nos dice que el modelo obtenido con bosques de decisión nos serviría mejor para realizar predicciones sobre los impuestos en las viviendas estudiadas.
